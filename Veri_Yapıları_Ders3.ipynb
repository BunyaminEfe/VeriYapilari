RECURSIVE Fonksiyonlar

Recursive fonksiyonlar, tanımlarında kendi kendini tekrar kullanan fonksiyonlardır. 

Bu fonksiyonlar, zor ve uzun problemleri daha kolay hallerine dönüştürerek adım adım çözmeyi sağlar.

```
def faktoriyel(x):
    if x == 0:
        return 1
    else:
        return x * faktoriyel(x-1)
```

```
faktoriyel(5)
# 1*2*3*4*5 =120
```

Bir de recursive olarak fibonacci dizisinin n. terimini bulma fonksiyonuna bakalım. Bu dizinin özelliği her bir terimin kendinden bir önceki ve iki önceki terimin toplamı olmasıdır.


1,1,2,3,5,8,13,21,34,.......

```
def fibonacci(n):
    if n <= 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

yield deyimi, return deyimi gibi fonksiyonlarda kullanılır, ancak, fonksiyon bir generator döndürür. 

```
def creategeneratorSquare(l):
  for x in l:
    yield x * x

generator = creategeneratorSquare([1,2,3,4,5])
for k in generator:
   print (k)
```

```
class Ogretmen:
  def __init__(self,isim,soyisim,yas,maas,uzmanlik):
    self.isim=isim
    self.soyisim=soyisim
    self.yas=yas
    self.maas=maas
    self.uzmanlik=uzmanlik
  
  def info(self):
    print("isim :{} soyadı : {} yas: {} uzmanlık alanı: {} maaş: {}"
    .format(self.isim,self.soyisim,self.yas,self.uzmanlik,self.maas))
  
  def zam(self):
    return self.maas*1.2
```

```
bir=Ogretmen("Gökhan","Tosun",27,3000,"Bilgisayar")
```

```
bir.info()
```

```
bir.zam()
```

```
class Araba():
  marka=""
  renk=""
  plaka=""
  hiz=0

  def hiziArttir(self):
    self.hiz+=10# self.hiz=self.hiz+10
    return self.hiz
```

```
araba1=Araba()
araba1.marka="Fiat"
araba1.renk="Beyaz"
araba1.plaka="34ABC34"
print("Marka {} Renk {} Plaka {} Hız {}"
.format(araba1.marka,araba1.renk,araba1.plaka,araba1.hiz))
araba1.hiziArttir()
print("Hız :",araba1.hiz)
print("Marka {} Renk {} Plaka {} Hız {}"
.format(araba1.marka,araba1.renk,araba1.plaka,araba1.hiz))
```
